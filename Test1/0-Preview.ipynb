```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '/home/corn/Documents/ScienceFair/Data/Data_1/Tri/t3.xlsx' # 路徑
data = np.array(pd.read_excel(path).values)
stop = next(i[1] for i in data if i[4] != 0)
data[:,1] = (data[:,1] - stop) / 15
```

Max Strees & Strain

```python
maxarg = np.argmax(data[:,4])

maxstrees = data[maxarg][4]
maxstrain = data[maxarg][5]

print("Max Strees :", maxstrees)
print("Max Strain :", maxstrain)
```

Preview Image

```python
plt.rcParams.update({'font.size': 10})
plt.figure(figsize=(3, 3), dpi=1000)

plt.plot(data[:,1], data[:,4]) # 1:mm 4:應力 5:應變

plt.ylabel("Strees (N/mm\u00b2)")
plt.xlabel("Strain (%)")

plt.tick_params(axis='x', direction='in')
plt.tick_params(axis='y', direction='in')

plt.show()
```
